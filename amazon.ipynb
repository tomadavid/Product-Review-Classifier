```python
import pandas as pd
import numpy as np
```

```python
# preprocessing a review
def preprocessing(text):
    #text = text.lower()
    #text = text.translate(str.maketrans('', '', string.punctuation))
    words = text.split()
    for word in words:
        if len(word) < 4:
            words.remove(word)
    #words = [word for word in words if word not in stopwords.words('english')]
    return words
```

```python
data = pd.read_csv('train.csv')
labels = data.iloc[:,0]
reviews = data.iloc[:,2]
```

```python
processed_reviews = [preprocessing(review) for review in reviews[:10000]]

processed_reviews
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences=processed_reviews, vector_size=100, window=5, min_count=1, workers=4)

vector = model.wv['best']
vector
```

```python
similar_words = model.wv.most_similar("greatest", topn=3)
print(similar_words)  # List of words similar to "machine"
```

```python
def average_review(review_mat):
    average = []
    word_count = 0
    
    for word in review_mat:
        if len(average) == 0:
            average = word
        else:
            average = np.add(average, word)
        word_count += 1

    average = average/word_count

    return average
```
